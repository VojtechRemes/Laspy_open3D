# Laspy + Open3D


## Co je `laspy`
`laspy` je Python knihovna pro čtení a práci se soubory **LAS/LAZ**, tedy standardními formáty pro ukládání lidarových a dalších bodových mračen.  
Pomůže nám:
- načíst data ze souboru,
- získat souřadnice bodů,
- pracovat s klasifikací bodů,
- číst atributy jako RGB barvu.

## Co je `open3d`
`open3d` je knihovna pro **3D geometrii a vizualizaci**.  
V tomto cvičení ji použijeme hlavně pro:
- vytvoření point cloudu,
- přiřazení barev bodům,
- interaktivní 3D zobrazení.
